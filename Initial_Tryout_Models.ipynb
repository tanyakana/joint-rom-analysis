```python
# prompt: process the data using Isolation Forest

from sklearn.ensemble import IsolationForest
df = pd.read_csv('Normal Dataset.csv')
df = df.drop('RACE', axis=1)
df = df.drop('GENDER', axis=1)
df_pred = df.copy()
# Assuming 'df' is your DataFrame and you want to detect anomalies in all columns
model = IsolationForest(contamination=0.01) # Adjust contamination as needed
model.fit(df_pred)

df['anomaly_score'] = model.decision_function(df_pred)
df['anomaly'] = model.predict(df_pred) # 1 for inliers, -1 for outliers

# Display the DataFrame with the new anomaly columns
df

#df['anomaly'].value_counts()
```

```python
# prompt: show me all the rows that have -1 anomaly

import pandas as pd
from sklearn.ensemble import IsolationForest

df = pd.read_csv('Normal Dataset.csv')

df = df.drop('RACE', axis=1)
df = df.drop('GENDER', axis=1)
df_pred = df.copy()
# Assuming 'df' is your DataFrame and you want to detect anomalies in all columns
model = IsolationForest(contamination=0.01) # Adjust contamination as needed
model.fit(df_pred)

df['anomaly_score'] = model.decision_function(df_pred)
df['anomaly'] = model.predict(df_pred) # 1 for inliers, -1 for outliers

# Display rows where 'anomaly' is -1
print(df[df['anomaly'] == -1])
```

```python
# prompt: need another way of detecting anomilies in the dataset using ml

import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
df = pd.read_csv('Normal Dataset.csv')
df = df.drop('RACE', axis=1)
df = df.drop('GENDER', axis=1)
df_pred = df.copy()
# Assuming 'df' is your DataFrame and you want to detect anomalies in all columns
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.01) # Adjust parameters as needed
df['anomaly_lof'] = lof.fit_predict(df_pred) # 1 for inliers, -1 for outliers

# Display the DataFrame with the new anomaly columns
df

#df['anomaly_lof'].value_counts()
```

```python
from sklearn.mixture import GaussianMixture
df = pd.read_csv('Normal Dataset.csv')
df = df.drop('RACE', axis=1)
df = df.drop('GENDER', axis=1)
df_pred = df.copy()
model = GaussianMixture(n_components=3)  # Adjust n_components as needed
model.fit(df_pred)
df['anomaly_gmm'] = model.score_samples(df_pred)  # Lower scores indicate anomalies
df['anomaly_gmm'].value_counts()
```

```python
from sklearn.svm import OneClassSVM
df = pd.read_csv('Normal Dataset.csv')
df = df.drop('RACE', axis=1)
df = df.drop('GENDER', axis=1)
df_pred = df.copy()
model = OneClassSVM(nu=0.01)  # Adjust nu as needed
model.fit(df_pred)  # Assuming 'df_pred' is your preprocessed DataFrame
df['anomaly_ocsvm'] = model.predict(df_pred)  # 1 for inliers, -1 for outliers
df['anomaly_ocsvm'].value_counts()
```
